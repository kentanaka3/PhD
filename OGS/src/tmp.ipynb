```python
import numpy as np
import pandas as pd
from pathlib import Path
from obspy import UTCDateTime
from datetime import datetime
from matplotlib import pyplot as plt

import ogsconstants as OGS_C
```

```python
waveforms = Path("/Volumes/Expansion/KEN/waveforms")
stations = Path("/Users/admin/Desktop/OGS_Catalog/station")
```

```python
start = datetime.strptime("20240320", OGS_C.YYYYMMDD_FMT)
end = datetime.strptime("20240620", OGS_C.YYYYMMDD_FMT)
waveforms, inventory = OGS_C.waveforms(
  waveforms,
  stations,
  start=start,
  end=end,
  output=Path("."),
)
```

```python
# Generate list of all dates in the range
DAYS = np.arange(start, end + OGS_C.ONE_DAY, OGS_C.ONE_DAY, # type: ignore
                 dtype='datetime64[D]').tolist() # type: ignore
DAYS = [UTCDateTime(day).date for day in DAYS]
Y = {day: {net: 0 for net in waveforms[OGS_C.NETWORK_STR].unique()}
     for day in DAYS}
for (date, net), df in waveforms.groupby([OGS_C.DATE_STR, OGS_C.NETWORK_STR]):
  Y[date][net] = len(df[OGS_C.STATION_STR].unique())
tmp = pd.DataFrame(Y).sort_index().T
tmp
```
